##1 Using def to import fasta sequences

```python
#First pull the information from the sequences file
def get_sequences_from_file(fasta_fn): 
    sequence_data_dict = {} #Creating an empty dictionary
    for record in SeqIO.parse(fasta_fn, "fasta"): #SeqIO.parse(fassta_fn, "fasta") reads the FASTA file (fasta_fn). The loop iterates over these records one by one.
        description = record.description.split() #Processing each record
        species_name = description[1] + " " + description[2]
        sequence_data_dict[species_name] = record.seq
    return(sequence_data_dict) #Returning the result        
```

##2 Function that translates a sting of nucleotides to amino acid.

```python
#Defining a funtion named 'translate_function' that takes one argument 'string_nucleotides' which represents a DNA seq string.
def translate_function(string_nuc): 
    mitoch_table = CodonTable.unambiguous_dna_by_name["Vertebrate Mitochondrial"] #Accessing genetic code    
    aa_string = [] # Initializing am Empty List
    for i in range(0, len(string_nuc),3): ##This loop iterates over the DNA sequence in steps of 3 and extracts each codon.
        codon = string_nuc[i:i +3]  
        
        if codon == 'AGA' or codon == 'AGG' or codon == 'TAG' or codon == 'TAA' : # looks for stop codons
            break ##If the current codon is a stop codon, the loop breaks        
        else:
            aa = mitoch_table.forward_table[codon] ##If the codon is not a stop codon, it translates the codon into an amino acid using the Vertebrate Mitochondrial genetic code and appends the amino acid to the 'amino_string'.
            aa_string.append(aa)     
       
    return "".join(aa_string) # Finally, this line join the amino acids in the 'amino_string' and return the translated amino acid sequence.
```

##3 Alternative translation function

```python
#Using Biopython's 'Seq.translate()' to translate the DNA sequence 'string_nucleotides' into amino acid sequence and uses to_stop = True to stop trabslation at stop codons. 
def translate(string_nuc): 
    amino_acid = Seq.translate(string_nuc, table="Vertebrate Mitochondrial", to_stop=True) 
    return str(amino_acid) 
```

##4 Function that calculates the molecular weight of each 3 amino acid sequence

```python
def calculate_MW(amino_acid_seq): 
    molecular_weight = ProteinAnalysis(amino_acid_seq).molecular_weight() #Using 'ProteinAnalysis' from Biopython to create an instance with amino acid sequence 'a_a_seq'. 'molecular_weight' to calculate molecular weight of the protein represented by the amino acid sequence.    
    return molecular_weight
```

##5 Function that computes the GC-content of each DNA sequence

```python
#get_sequences_from_file' function read the FASTA file and return a dictionary containing species name. The return dictionary is assigned to the variable 
cytb_seqs = get_sequences_from_file("penguins_cytb.fasta") 
penguins_df = pd.read_csv("penguins_mass.csv") #Using pandas('pd') to read a CSV file named "penguins_mass.csv".
species_list = list(penguins_df.species) ##This line creates a list named 'species_list' containing unique values from the species column of the Dataframe 'penguine_df'.
```

```python
def GC_content(DNA_string):
    G_count = DNA_string.count("G") #This line counts the occurrences of the character "G" in the 'DNA_string' and store the count in the variable 'G_count'.    
    C_count = DNA_string.count("C") ##This line counts the occurrences of the character "C" in the 'DNA_string' and store the count in the variable 'C_count'.    
    DNAstring_length = len(DNA_string) #This line calculates the GC content by adding the counts of "C" and "G" and then dividing by the total length of the DNA string. The result is stored in the variable 'GC_content'.
    GC_content = (C_count+G_count)/DNAstring_length 
    return GC_content #Finally, return the calculated GC content value.
```

##6 Adding two new columns to the penguin DataFrame: (1) molecular weight and (2) GC content.

```python
#Creating a new DataFrame named 'penguins_column_extended' by assigning two new columns to the existing 'penguins_df'. The assign() is used to add these two columns.
penguins_column_extended = penguins_df.assign(molecular_weight = 'NaN', GC_content='NaN') # adds the columns to the table
print(penguins_column_extended)
```

##7 Calling functions from step 3 (or step 2) and step 4 and fill in the new columns in the DataFrame.

```python
#Starting loop that iterates over each item in the 'cytb_seqs'. 
for key, value in cytb_seqs.items(): 
    aa_seq = translate_function(value) #'translate_fuction translates the DNA seq into an amino acid seq'
    molecular_weight = calculate_MW(aa_seq) #Calculate the sum of the molecular weights of each amino acid
    GC_percentage = GC_content(value)*100 #Calculate the GC percentage which is multiplied by 100 to convert to percentage.
    
    for index, row in penguins_column_extended.iterrows(): #Another loop that iterates each row in 'penguins_column_extended'
        if row['species'] == key: #Checks if species column matches the current key from 'cybt_seqs'
            penguins_column_extended.loc[index, 'molecular_weight'] = molecular_weight #If species name matches then molecular weight calculated earlier is assigned.
            penguins_column_extended.loc[index, 'GC_content'] = GC_percentage #Similar for GC percentage.
            
```

##8 Plotting a bar-chart of adult body mass per species.

```python
#Importing 'matplotlib' library's pyplot as 'plt' and 'seaborn' library as 'sns'
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

colors = sns.color_palette('Set2', n_colors=len(penguins_column_extended)) #Generate list of color using Seaborn's color_palette function.

#Creating bar plot using 'penguins_column_extended' dataframe. 'Species' column is used for the x-axis and the 'mass' column is used for the y-axis.
penguins_column_extended.plot(kind='bar', x='species', y='mass', title='Mass of Species', color=colors)
plt.xlabel('Species') #Setting the label for x-axis 'Species'
plt.ylabel('Mass') #Setting the label for y-axis 'mass'
plt.xticks(rotation=45, ha='right') #Rotating the x-axis label to 45 degree
plt.tight_layout() #Adjusting parameters to avoid overlapping
plt.show()

```
